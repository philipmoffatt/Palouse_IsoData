### Isotope Data - Meterological Data Regression Analysis

```r
monthlyData<- read.csv('processed_data/Iso_Met_monthly.csv')
dailyData<- read.csv('processed_data/Iso_Met_daily.csv')
```

```r
plot(monthlyData[,c(12,13,6,19,16)])
plot(dailyData[,c(13,14,7,37,30)])
```

```r
gen.regression <- lm(X2H ~ tavg + rhmax +Longitude+ Elevation_m.x, monthlyData)
summary(gen.regression)
plot(gen.regression)
```

```r
gen.regression <- lm(X2H ~ tavg + rhmax +Longitude+ Elevation_m.x, dailyData)
summary(gen.regression)
plot(gen.regression)
```
